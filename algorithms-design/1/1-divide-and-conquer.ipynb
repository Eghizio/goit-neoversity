```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    
    # Recursive division
    left = merge_sort(left)
    right = merge_sort(right)
    
    return merge(left, right)

# Combine
def merge(left, right):
  result = []
  i, j = 0, 0
   
  while i < len(left) and j < len(right):
    print("- Sorting...")
    if left[i] < right[j]:
      result.append(left[i])
      i += 1
    else:
      result.append(right[j])
      j += 1
   
  result.extend(left[i:])
  result.extend(right[j:])
   
  return result
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            print(f"- Iteration {i+1}:{j+1}")
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

def bubble_sort_optimized(arr):
    n = len(arr)
    for i in range(n):
        swapped = False

        for j in range(0, n-i-1):
            print(f"- Iteration {i+1}:{j+1}")
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                
                swapped = True

        if not swapped:
            break
    return arr
```

```python
print("\nBubble sort:")
print(bubble_sort([38, 27, 43, 3, 9, 82, 10])) # O(n^2)

print("\nBubble sort optimized:")
print(bubble_sort_optimized([38, 27, 43, 3, 9, 82, 10])) # O(n^2)'

print("\nMerge sort:")
print(merge_sort([38, 27, 43, 3, 9, 82, 10])) # O(n log n)
```

```python
def quick_sort(arr):
    print("- Sorting...")
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

print("\nQuick sort:")
print(quick_sort([38, 27, 43, 3, 9, 82, 10])) # O(n log n)
```

```python
def binary_search(arr, target, low=0, high=None):
    print("- Searching...")
    if high is None:
        high = len(arr) - 1
    if low > high:
        return -1

    mid = (low + high) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] > target:
        return binary_search(arr, target, low, mid-1)
    else:
        return binary_search(arr, target, mid+1, high)

print("\nBinary search: 3 (first)")
print(binary_search([3, 9, 10, 27, 38, 43, 82], 3)) # O(log n)

print("\nBinary search: 27 (midddle)")
print(binary_search([3, 9, 10, 27, 38, 43, 82], 27)) # O(1)

print("\nBinary search: 82 (last)")
print(binary_search([3, 9, 10, 27, 38, 43, 82], 82)) # O(log n)
```

```python
DEBUG = False
def debug_print(message):
    if DEBUG: print(message)

def max_crossing_sum(arr, left, mid, right):
    debug_print("- Crossing sum...")
    
    left_sum = float('-inf')
    total = 0
    max_left = mid
    for i in range(mid, left-1, -1):
        debug_print("- Crossing sum first loop...")
        total += arr[i]
        if total > left_sum:
            left_sum = total
            max_left = i
    
    right_sum = float('-inf')
    total = 0
    max_right = mid + 1
    for i in range(mid + 1, right + 1):
        debug_print("- Crossing sum second loop...")
        total += arr[i]
        if total > right_sum:
            right_sum = total
            max_right = i
    
    return left_sum + right_sum, max_left, max_right

def max_subarray_sum(arr, left, right):
    debug_print("- Sub array sum...")
    if left == right:
        return arr[left], left, right
    
    mid = (left + right) // 2
    
    left_sum, left_start, left_end = max_subarray_sum(arr, left, mid)
    right_sum, right_start, right_end = max_subarray_sum(arr, mid + 1, right)
    cross_sum, cross_start, cross_end = max_crossing_sum(arr, left, mid, right)
    
    if left_sum >= right_sum and left_sum >= cross_sum:
        return left_sum, left_start, left_end
    elif right_sum >= left_sum and right_sum >= cross_sum:
        return right_sum, right_start, right_end
    else:
        return cross_sum, cross_start, cross_end


arr = [2, -4, 1, 9, -6, 7, -3]
max_sum, start, end = max_subarray_sum(arr, 0, len(arr)-1) # O(n log n)
print(f"\nSubarray: {arr[start:end+1]} Sum: {max_sum}")
```
