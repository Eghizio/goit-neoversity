```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from dataclasses import dataclass
from typing import List

@dataclass
class Fn:
    """Function arguments and values with Plot label."""
    x: List[int]
    y: List[int]
    label: str

@dataclass
class PlotOptions:
    """Plot options."""
    title: str
    xlabel: str
    ylabel: str
    logarithmic: bool = False

def plot_fns(fns: List[Fn], options: PlotOptions = None):
    if options == None:
        options = PlotOptions("Algorithms time complexity", "n", "Operations", logarithmic=False)

    plt.figure(figsize=(12, 8))

    if options.logarithmic:
        plt.yscale("log")

    for fn in fns:
        plt.plot(fn.x, fn.y, label=fn.label)

    plt.title(options.title)
    plt.xlabel(options.xlabel)
    plt.ylabel(options.ylabel)

    plt.legend()
    plt.grid(True, which="both", ls="--", c='0.65')
    
    plt.show()
```

```python
N = 10

n = np.arange(1, N)

O_1 = np.ones_like(n)
O_n = n
O_n_squared = n**2
O_n_cubed = n**3

print(n, O_1, O_n, O_n_squared, O_n_cubed)

plot_fns(
    [
        Fn(n, O_1, "O(1)"),
        Fn(n, O_n, "O(n)"),
        Fn(n, O_n_squared, "O(n^2)"),
        Fn(n, O_n_cubed, "O(n^3)"),
    ],
    # PlotOptions("Algorithms time complexity", "n", "Operations", logarithmic=False)
    PlotOptions("Algorithms time complexity", "n", "Operations", logarithmic=True)
)

```

```python
# f(n)=3n^2 + 2n + 1
def f(n): return 3*(n**2) + 2*n + 1
def big_o(n): return 4*(n**2)
def big_omega(n): return 3*(n**2)

plot_fns([
    Fn(n, f(n), "f(n)"),
    Fn(n, big_o(n), "O(n) - Worst case scenario"),
    Fn(n, big_omega(n), "Omega(n) - Best case scenario"),
])
```

```python
def get_element(arr, index):
    print(f"- Getting element: {index}")
    return arr[index]

def lookup(hash_map, key):
    print(f"- Looking up item: {key}")
    return hash_map[key]

print("\nGet list element by index: 3")
print(get_element([1, 2, 3, 42], 3)) # O(1)

print("\nGet hash map value by key: 'admin'")
print(lookup({ "user": "Adam", "admin": "Kuba" }, "admin")) # O(1)
```

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        print(f"- Iteration {i+1}")

        if arr[i] == target:
            return i
    return -1

arr = [1, 2, 3, 4, 5]

print("\nLinear search: 1 (first)")
linear_search(arr, 1) # O(1)

print("\nLinear search: 3 (middle)")
linear_search(arr, 3) # O(n)

print("\nLinear search: 5 (last)")
linear_search(arr, 5) # O(n)
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            print(f"- Iteration {i+1}:{j+1}")
            if arr[j] > arr[j+1]:
                print("Swapping elements...")
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

print("\nBubble sort: Sorted")
bubble_sort([1, 2, 3, 4, 5])

print("\nBubble sort: Unsorted")
bubble_sort([5, 4, 3, 2, 1])
```

```python
def bubble_sort_optimized(arr):
    n = len(arr)
    for i in range(n):
        swapped = False

        for j in range(0, n-i-1):
            print(f"- Iteration {i+1}:{j+1}")
            if arr[j] > arr[j+1]:
                print("Swapping elements...")
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True

        if not swapped:
            break
    return arr

print("\nBubble sort optimized: Sorted")
bubble_sort_optimized([1, 2, 3, 4, 5]) # O(n)

print("\nBubble sort optimized: Unsorted")
bubble_sort_optimized([5, 4, 3, 2, 1]) # O(n^2)
```

```python
def sum_array(arr):
    total = 0
    for i, x in enumerate(arr):
        print(f"- Iteration {i+1}")
        total += x
    return total

sum_array([1, 2, 3, 4, 5]) # O(n)
```

```python
def find_duplicate_pairs(arr):
    duplicates = []
    n = len(arr)
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] == arr[j]:
                # duplicates.append((arr[i], [i, j]))
                duplicates.append(arr[i])
    return duplicates

find_duplicate_pairs([1, 2, 3, 4, 5, 3, 2]) # O(n^2)
```

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        print("- Searching...")
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

print("\nBinary search: 1 (first)")
binary_search([1, 2, 3, 4, 5, 6, 7, 8, 9], 1) # O(log n)

print("\nBinary search: 5 (middle)")
binary_search([1, 2, 3, 4, 5, 6, 7, 8, 9], 5) # O(1)

print("\nBinary search: 9 (last)")
binary_search([1, 2, 3, 4, 5, 6, 7, 8, 9], 9) # O(log n)
```
