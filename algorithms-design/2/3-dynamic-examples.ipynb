```python
# Levenshtein distance
def edit_distance(str1: str, str2: str) -> int:
    m, n = len(str1), len(str2)
    
    # Create a table to store subproblem results
    dp = [[0 for _ in range(n + 1)] for _ in range(m + 1)]
    
    # Fill the first column and first row
    for i in range(m + 1):
        dp[i][0] = i
    for j in range(n + 1):
        dp[0][j] = j
    
    # Fill the rest of the table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if str1[i-1] == str2[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                dp[i][j] = 1 + min(dp[i-1][j],      # deletion
                                   dp[i][j-1],      # insertion
                                   dp[i-1][j-1])    # substitution
    
    return dp[m][n]

# Example usage
str1 = "kitten"
str2 = "sitting"
print(f"Edit distance between '{str1}' and '{str2}': {edit_distance(str1, str2)}")
```

```python
# LIS - Longest Increasing Subsequence
def longest_increasing_subsequence(arr):
    n = len(arr)
    # Initialise the dp array, where dp[i] represents the length of the LIS ending at index i
    dp = [1] * n
    
    # Fill the dp array
    for i in range(1, n):
        for j in range(i):
            if arr[i] > arr[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    
    # Find the maximum value in dp
    max_length = max(dp)
    
    # Reconstruct the actual sequence
    sequence = []
    max_index = dp.index(max_length)
    sequence.append(arr[max_index])
    for i in range(max_index - 1, -1, -1):
        if dp[i] == max_length - 1 and arr[i] < arr[max_index]:
            sequence.append(arr[i])
            max_length -= 1
            max_index = i
    
    return list(reversed(sequence))

# Example usage
arr = [10, 22, 9, 33, 21, 50, 41, 60, 80]
lis = longest_increasing_subsequence(arr)
print(f"Longest Increasing Subsequence: {lis}")
print(f"Length: {len(lis)}")
```

```python
# Kadane Algorithm

def max_subarray_sum(arr):
    n = len(arr)
    max_sum = float('-inf')
    current_sum = 0
    start = 0
    end = 0
    temp_start = 0

    for i in range(n):
        current_sum += arr[i]

        if current_sum > max_sum:
            max_sum = current_sum
            start = temp_start
            end = i

        if current_sum < 0:
            current_sum = 0
            temp_start = i + 1

    return max_sum, arr[start:end+1]

# Example of using
arr = [2, -4, 1, 9, -6, 7, -3]
max_sum, subarray = max_subarray_sum(arr)
print(f"Maximum subarray sum: {max_sum}")
print(f"Subarray with the maximum sum: {subarray}")
```
